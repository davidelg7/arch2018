```python
import matplotlib.pyplot as plt
import pandas as pd
r=80
fig =plt.figure(num=None, figsize=(15, 15),dpi=100, facecolor='w', edgecolor='k')
sub=fig.add_subplot(3,2,1)
p1 = pd.read_csv("p 0 c 0 g 0.csv")
p2= pd.read_csv("p 0 c 1 g 0.csv")
x1 = p1["x"]
y1= p1["y"]
x2 = p2["x"]
y2= p2["y"]
sub.set_title("Passo 0")

sub.set_xlabel("x")
sub.set_ylabel("y")
sub.scatter(x1[1:],y1[1:], color='g', s=r, alpha=0.7)
sub.scatter(x1[0],y1[0], color='r',s=80) 

sub.scatter(x2[1:],y2[1:], color='y',s=r, alpha=0.7)
sub.scatter(x2[0],y2[0], color='b',s=80)
sub.grid(color='grey', linestyle='-', linewidth=0.5, alpha=0.5)

sub.legend(["Punti mappati da C0","C0 casuale","Punti mappati da C1","C1 casuale"])


for p in range(1,6):
    sub=fig.add_subplot(3,2,p+1)
    p1 = pd.read_csv("p "+str(p)+" c 0 g 0.csv")
    p2= pd.read_csv("p "+str(p)+" c 1 g 0.csv")
    oldc0x=x1[0]
    oldc0y=y1[0]
    oldc1x=x2[0]
    oldc1y=y2[0]
    x1 = p1["x"]
    y1= p1["y"]
    x2 = p2["x"]
    y2= p2["y"]
    sub.set_title("Passo "+str(p))
    sub.scatter(oldc0x,oldc0y,color='r', s=80,alpha=0.8)
    sub.scatter(oldc1x,oldc1y,color='b', s=80,alpha=0.8)
    plt.annotate("", xy=(x1[0],y1[0] ), xytext=(oldc0x, oldc0y), arrowprops=dict(arrowstyle="->",color='r'))
    plt.annotate("", xy=(x2[0],y2[0] ), xytext=(oldc1x, oldc1y), arrowprops=dict(arrowstyle="->",color='b'))
    
    sub.scatter(x1[1:],y1[1:], color='g',s=r, alpha=0.7)
    sub.scatter(x1[0],y1[0], color='r',s=80) 

    sub.scatter(x2[1:],y2[1:], color='y',s=r, alpha=0.7)
    sub.scatter(x2[0],y2[0], color='b',s=80)

    sub.set_xlabel("x")
    sub.set_ylabel("y")
    sub.legend(["C0 vecchio","C1 vecchio","C0","C1","Punti mappati da C0","Punti mappati da C1"])
    sub.grid(color='grey', linestyle='-', linewidth=0.5, alpha=0.5)

plt.plot()
fig.savefig("Test.pdf", bbox_inches='tight')
```
