```python
import matplotlib.pyplot as plt
import pandas as pd
fig =plt.figure(num=None, figsize=(20, 16), dpi=100, facecolor='w', edgecolor='k')

for p in range(9):
    sub=fig.add_subplot(3,3,p+1)
    p1 = pd.read_csv("p "+str(p)+" c 0 g 0.csv")
    p2= pd.read_csv("p "+str(p)+" c 1 g 0.csv")
    x1 = p1["x"]
    y1= p1["y"]
    x2 = p2["x"]
    y2= p2["y"]
    
    sub.scatter(x1[0],y1[0], color='r') 
    plt.annotate("Centroide 0", xy=(x1[0], y1[0]), xytext=(x1[0], max(y1)+1), arrowprops=dict(arrowstyle="->"))
    sub.scatter(x1[1:],y1[1:], color='g')
    
    sub.scatter(x2[0],y2[0], color='b')
    plt.annotate("Centroide 1", xy=(x2[0], y2[0]), xytext=(x2[0], max(y2)+1), arrowprops=dict(arrowstyle="->"))
    sub.scatter(x2[1:],y2[1:], color='y')

plt.plot()
fig.savefig("Test.pdf", bbox_inches='tight')
```
